```python
import numpy as np
import pandas as pd
import scipy.stats as stats
import matplotlib.pyplot as plt

# Set plotting options
%matplotlib inline
plt.rc('figure', figsize=(16, 9))
```

```python
# Sample A: Standard Normal distribution
sample_a = stats.norm.rvs(loc=0.0, scale=1.0, size=(1000,))

# Sample B: Non-standard-normal distribution
sample_b = stats.norm.rvs(loc=10.0, scale=20.0, size=(1000,))
```

```python
sorted_a = np.sort(sample_a)
sorted_b = np.sort(sample_b)

plt.plot(sorted_a, sorted_b, 'o')
```

```python
sorted_b_normalized = (sorted_b - np.mean(sorted_b)) / np.std(sorted_b)

plt.plot(sorted_a, sorted_b_normalized, 'o')
```

```python
# Sample B: Non-normal distribution
sample_log_norm = stats.lognorm.rvs(s=0.5, loc=20.0, scale=50.0, size=(1000,))
sorted_log_norm = np.sort(sample_log_norm)

plt.plot(sorted_a, sorted_log_norm, 'o')
```

```python
sorted_log_norm_normalized = (sorted_log_norm - np.mean(sorted_log_norm)) / np.std(sorted_log_norm)

plt.plot(sorted_a, sorted_log_norm_normalized, 'o')
```
